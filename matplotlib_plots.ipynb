```python
import matplotlib.pyplot as plt

x = ['a','b','c']
y = [10, 12, 8]

plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [5, 7, 4, 6, 9]

plt.scatter(x, y, color='black')
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['A', 'B', 'C', 'D']
y = [5, 7, 3, 8]

plt.bar(x, y, color='green')
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [1, 1, 2, 3, 3, 3, 4, 5, 5, 5, 6, 6, 7, 8]

plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

names = ['Python', 'C++', 'Java', 'JavaScript','html']
sizes = [40, 25, 20, 10,5]

plt.pie(sizes, labels=names, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart Example")
plt.axis('equal')  # Makes the pie chart circular
plt.show()
```
